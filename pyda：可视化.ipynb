# wtfis df.asof()?

>DataFrame.asof(self, where, subset=None)[source]
<br>Return the last row(s) without any NaNs before where.
><br>The last row (for each element in where, if list) without any NaN is taken. In case of a DataFrame, the last row <br>without NaN considering only the subset of columns (if not None)

><br>New in version 0.19.0: For DataFrame
><br>If there is no good value, NaN is returned for a Series or a Series of NaN values for a DataFrame


```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('_01_pyda_reference_2nd_Edition/examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy'),
    (datetime(2009, 3, 5), 'bb')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 150), xytext=(date, spx.asof(date) + 300),
                arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1900])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```
