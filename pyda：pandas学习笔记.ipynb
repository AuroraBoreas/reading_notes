```python
# mnames = ['movie_id', 'title', 'genres']
movies = pd.read_csv('_01_pyda_reference_2nd_Edition/datasets/movielens/movies.dat',header=None, sep=r'::', engine='python')
movies = movies.rename({0:'movie_id',1:'title',2:'genre'},axis='columns')
movies[:10]
```

```python
genre = movies['genre']
genre[0].split('|')
```

```python
genre = []
for i in movies['genre']:
    genre.extend(i.split('|'))
genre
```

```python
genre = pd.unique(genre)
genre
```

```python
zero_matrix = np.zeros((len(movies), len(genre)))
dummies = pd.DataFrame(zero_matrix, columns=genre)
dummies[:10]
```

```python
test = movies['genre']
test[0].split('|')

indexer = dummies.columns.get_indexer(test[0].split('|'))
dummies.iloc[0, indexer]=1
dummies[:10]
```

```python
for i, gen in enumerate(movies.genre):
    indexer = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indexer] = 1
```

```python
dummies[:10]
```

```python
new_movies = movies.join(dummies.add_prefix('gen_'))
new_movies.head(5)
```

# 理解Index.get_indexer()方法

# 生成随机样本，随机采样
1. pd.get_dummies() + df.take() + np.random.permutation(n)
2. df.sample(n)

```python
np.random.seed(12345)
samples = np.random.randint(0, 1000, size=(200,5))
df = pd.DataFrame(samples, columns=np.arange(5))
df = df.add_prefix('h')
df.head()
```

## 采样

```python
new_samples = df.sample(n=10)
new_samples
```

```python
indexs = np.random.randint(0,200, size=10) #随机从0-200中抽出10个行号
sampler = np.random.permutation(indexs)  #随机排列这10个行号
re_samples = df.take(sampler)#抽出结果
re_samples
```

```python
newnew_samples = re_samples['h0']
bins = [0, 220, 420, 620, 820, 1200]
cuts = pd.cut(newnew_samples, bins)
cuts.value_counts()
```

```python
new_dummies = pd.get_dummies(re_samples['h0'], prefix='x_')
new_dummies
```
